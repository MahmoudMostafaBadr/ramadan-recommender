```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 300)
pd.set_option("display.width", 200)

df = pd.read_csv("epi_r.csv")   

print("Shape:", df.shape)
df.head()
```

```python
core_cols = ["title", "rating", "calories", "protein", "fat", "sodium"]

tag_cols = [
    "ramadan", "healthy", "breakfast", "dinner", "lunch", "snack",
    "low sugar", "low sodium", "high fiber", "low fat", "low carb",
    "vegetarian", "vegan", "kidney friendly"
]

tag_cols = [c for c in tag_cols if c in df.columns]

recipes = df[core_cols + tag_cols].copy()

print("Before cleaning:", recipes.shape)

recipes = recipes.dropna(subset=["title", "calories", "protein", "fat", "sodium"])


for c in ["rating", "calories", "protein", "fat", "sodium"]:
    recipes[c] = pd.to_numeric(recipes[c], errors="coerce")

recipes = recipes.dropna(subset=["calories", "protein", "fat", "sodium"])

print("After cleaning:", recipes.shape)
recipes.head()
```

```python
recipes = recipes.copy()

# 1) satiation score (شبع) — heuristic
# البروتين يشبع، الدهون تشبع بس لو زادت تقلب تقيلة، الصوديوم يعطش
recipes["satiation_score"] = (
    0.7 * recipes["protein"] +
    0.2 * recipes["fat"] -
    0.3 * (recipes["sodium"] / 1000)
)

# 2) hydration penalty (عطش)
recipes["hydration_penalty"] = recipes["sodium"] / 1000

# 3) meal_slot rules
def assign_meal_slot(row):
    cal = row["calories"]
    prot = row["protein"]
    fat = row["fat"]
    sod = row["sodium"]

    # suhoor: بروتين أعلى + دهون معتدلة + صوديوم أقل
    if (prot >= 20) and (fat <= 35) and (sod <= 900) and (cal <= 800):
        return "suhoor"

    # iftar: سعرات متوسطة + مش تقيلة قوي
    if (cal >= 250) and (cal <= 900) and (fat <= 45) and (sod <= 1200):
        return "iftar"

    return "either"

recipes["meal_slot"] = recipes.apply(assign_meal_slot, axis=1)

recipes["meal_slot"].value_counts()
```

```python
clean["satiation_score"] = (
    0.7 * clean["protein"] +
    0.2 * clean["fat"] -
    0.3 * (clean["sodium"] / 1000)
)

def assign_meal_slot(row):
    cal = row["calories"]
    prot = row["protein"]
    fat = row["fat"]
    sod = row["sodium"]

    if (prot >= 20) and (fat <= 35) and (sod <= 900) and (cal <= 800):
        return "suhoor"
    if (cal >= 250) and (cal <= 900) and (fat <= 45) and (sod <= 1200):
        return "iftar"
    return "either"

clean["meal_slot"] = clean.apply(assign_meal_slot, axis=1)

clean["meal_slot"].value_counts()
```

```python
def recommend(data, meal="suhoor", 
              calories_max=800, 
              protein_min=20, 
              sodium_max=1200, 
              top_n=10):
    
    # 1) فلترة أولية حسب قيود المستخدم
    filtered = data[
        (data["meal_slot"] == meal) &
        (data["calories"] <= calories_max) &
        (data["protein"] >= protein_min) &
        (data["sodium"] <= sodium_max)
    ].copy()
    
    if filtered.empty:
        return "No recipes found with these constraints."
    
    # 2) ترتيب حسب satiation_score + rating
    filtered["final_score"] = (
        0.7 * filtered["satiation_score"] +
        0.3 * filtered["rating"]
    )
    
    results = filtered.sort_values("final_score", ascending=False).head(top_n)
    
    return results[[
        "title", 
        "meal_slot", 
        "calories", 
        "protein", 
        "fat", 
        "sodium", 
        "final_score"
    ]]
```

```python
import pandas as pd
import numpy as np

def diversify_topn(df_ranked, top_n=10, group_cols=None, per_group_cap=3):
    """
    df_ranked: DataFrame مرتب تنازليًا على final_score
    group_cols: أعمدة tags نستخدمها لتحديد نوع الوصفة (0/1)
    per_group_cap: أقصى عدد من كل مجموعة داخل Top-N
    """
    if group_cols is None:
        group_cols = []

    chosen = []
    group_counts = {g: 0 for g in group_cols}

    for _, row in df_ranked.iterrows():
        # تحديد المجموعة الأولى اللي الوصفة بتنتمي لها
        group = None
        for g in group_cols:
            if g in df_ranked.columns and row.get(g, 0) == 1:
                group = g
                break

        # لو مفيش مجموعة واضحة
        if group is None:
            group = "other"
            if "other" not in group_counts:
                group_counts["other"] = 0

        # تطبيق cap
        if group_counts[group] >= per_group_cap:
            continue

        chosen.append(row)
        group_counts[group] += 1

        if len(chosen) >= top_n:
            break

    if not chosen:
        return df_ranked.head(top_n)

    return pd.DataFrame(chosen)


def recommend_diverse(
    data,
    meal="suhoor",
    calories_max=800,
    protein_min=20,
    sodium_max=1200,
    top_n=10,
    pool_size=200,
    per_group_cap=3
):
    # 1) فلترة قيود المستخدم
    filtered = data[
        (data["meal_slot"] == meal) &
        (data["calories"] <= calories_max) &
        (data["protein"] >= protein_min) &
        (data["sodium"] <= sodium_max)
    ].copy()

    if filtered.empty:
        return filtered  # فاضي

    # 2) Ranking
    filtered["rating_filled"] = filtered["rating"].fillna(0)
    filtered["final_score"] = (
        0.75 * filtered["satiation_score"] +
        0.25 * filtered["rating_filled"]
    )

    # 3) ناخد pool كبير
    ranked = filtered.sort_values("final_score", ascending=False).head(pool_size)

    # 4) إزالة التكرار حسب العنوان
    ranked = ranked.drop_duplicates(subset=["title"], keep="first")

    # 5) مجموعات للتنوع (لو موجودة)
    group_cols = [c for c in ["seafood", "poultry", "vegetarian", "vegan", "beef", "lamb", "fish", "chicken"]
                  if c in ranked.columns]

    diverse = diversify_topn(ranked, top_n=top_n, group_cols=group_cols, per_group_cap=per_group_cap)

    # 6) Output + تفسير بسيط (إصلاح مشكلة dtype)
    out = diverse[["title", "meal_slot", "calories", "protein", "fat", "sodium", "final_score"]].copy()

    why1 = np.where(out["sodium"].astype(float) <= 600,
                    "Lower sodium (better hydration)",
                    "Higher sodium (watch hydration)")

    why2 = np.where(out["protein"].astype(float) >= 25,
                    "High protein",
                    "Moderate protein")

    out["why"] = (pd.Series(why1, index=out.index).astype(str)
                  + " + " +
                  pd.Series(why2, index=out.index).astype(str))

    return out.sort_values("final_score", ascending=False)
```

```python
import re

def infer_kind_from_title(title: str) -> str:
    t = str(title).lower()

    # seafood / fish
    if re.search(r"\b(salmon|tuna|shrimp|prawn|crab|lobster|clam|mussel|oyster|scallop|halibut|bass|cod|tilapia|trout|sardine|anchovy|fish)\b", t):
        return "seafood"

    # poultry
    if re.search(r"\b(chicken|turkey|duck|poultry)\b", t):
        return "poultry"

    # beef / lamb / veal
    if re.search(r"\b(beef|steak)\b", t):
        return "beef"
    if re.search(r"\b(lamb)\b", t):
        return "lamb"
    if re.search(r"\b(veal)\b", t):
        return "veal"

    # vegetarian-ish keywords
    if re.search(r"\b(vegetable|zucchini|eggplant|lentil|beans|chickpea|tofu|mushroom|spinach|salad)\b", t):
        return "veg"

    return "other"


def recommend_diverse_v2(
    data,
    meal="suhoor",
    calories_max=800,
    protein_min=20,
    sodium_max=1200,
    top_n=10,
    pool_size=300,
    per_kind_cap=2
):
    filtered = data[
        (data["meal_slot"] == meal) &
        (data["calories"] <= calories_max) &
        (data["protein"] >= protein_min) &
        (data["sodium"] <= sodium_max)
    ].copy()

    if filtered.empty:
        return filtered

    filtered["rating_filled"] = filtered["rating"].fillna(0)
    filtered["final_score"] = 0.75 * filtered["satiation_score"] + 0.25 * filtered["rating_filled"]

    ranked = filtered.sort_values("final_score", ascending=False).head(pool_size)
    ranked = ranked.drop_duplicates(subset=["title"], keep="first")

    # kind from title
    ranked["kind"] = ranked["title"].apply(infer_kind_from_title)

    # diversify by kind
    chosen = []
    counts = {}

    for _, row in ranked.iterrows():
        k = row["kind"]
        counts[k] = counts.get(k, 0)
        if counts[k] >= per_kind_cap:
            continue
        chosen.append(row)
        counts[k] += 1
        if len(chosen) >= top_n:
            break

    out = pd.DataFrame(chosen)
    if out.empty:
        out = ranked.head(top_n).copy()

    out = out[["title","kind","meal_slot","calories","protein","fat","sodium","final_score"]].copy()

    why1 = np.where(out["sodium"].astype(float) <= 600, "Lower sodium", "Higher sodium")
    why2 = np.where(out["protein"].astype(float) >= 25, "High protein", "Moderate protein")
    out["why"] = pd.Series(why1, index=out.index).astype(str) + " + " + pd.Series(why2, index=out.index).astype(str)

    return out.sort_values("final_score", ascending=False)
```

```python
recommend_diverse_v2(clean2, meal="suhoor", calories_max=700, protein_min=20, sodium_max=1200, top_n=10, pool_size=300, per_kind_cap=2)
```
